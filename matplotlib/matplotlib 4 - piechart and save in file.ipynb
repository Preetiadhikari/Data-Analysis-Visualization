```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals =[1400,600,300,410,250]
exp_labels =["home_rent","food","phone/interent bill","car","other utlities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,autopct='%0.0f%%',shadow=True,explode=[0,0,0,0.4,0],startangle=45)

plt.savefig("piechart.png")
plt.show()
```
